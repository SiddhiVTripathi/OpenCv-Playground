# Color Filtering

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    #hsv is hue sat value
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_blu = np.array([100,20,0])
    upper_blu = np.array([150,255,255])
    
    #dark_red = np.unit8([[[12,22,121]]])
    #dark_red = cv2.cvtColor(dark_red, cv2.COLOR_BGR2HSV)
    
    mask = cv2.inRange(hsv, lower_blu, upper_blu)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    #cv2.imshow('frame', frame)
    #cv2.imshow('mask', mask)
    cv2.imshow('res', res)
    
    k = cv2.waitKey(5) & 0xFF
    if k==ord("q"):
        break

cv2.destroyAllWindows()
cap.release()
```

## Smoothing blurring

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    #hsv is hue sat value
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_blu = np.array([100,20,0])
    upper_blu = np.array([150,255,255])
    
    #dark_red = np.unit8([[[12,22,121]]])
    #dark_red = cv2.cvtColor(dark_red, cv2.COLOR_BGR2HSV)
    
    mask = cv2.inRange(hsv, lower_blu, upper_blu)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    
    kernel = np.ones((5,5), np.float32)/25
    smoothed = cv2.filter2D(res, -1, kernel)
    #cv2.imshow('frame', frame)
    cv2.imshow('smoothed', smoothed)
    
    k = cv2.waitKey(5) & 0xFF
    if k==ord("q"):
        break

cv2.destroyAllWindows()
cap.release()
```

## Morphological Transformations

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    #hsv is hue sat value
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_blu = np.array([100,20,0])
    upper_blu = np.array([150,255,255])
    
    #dark_red = np.unit8([[[12,22,121]]])
    #dark_red = cv2.cvtColor(dark_red, cv2.COLOR_BGR2HSV)
    
    mask = cv2.inRange(hsv, lower_blu, upper_blu)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    kernel = np.ones((5,5), np.uint8)
    #erosion = cv2.erode(mask, kernel, iterations=1)
    dilation = cv2.dilate(mask, kernel, iterations=1)
    cv2.imshow('dilate', dilation)
    k = cv2.waitKey(5) & 0xFF
    if k==ord("q"):
        break

cv2.destroyAllWindows()
cap.release()
```

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    #hsv is hue sat value
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    lower_blu = np.array([100,20,0])
    upper_blu = np.array([150,255,255])
    
    #dark_red = np.unit8([[[12,22,121]]])
    #dark_red = cv2.cvtColor(dark_red, cv2.COLOR_BGR2HSV)
    
    mask = cv2.inRange(hsv, lower_blu, upper_blu)
    res = cv2.bitwise_and(frame, frame, mask=mask)
    
    kernel = np.ones((5,5), np.uint8)
    erosion = cv2.erode(mask, kernel, iterations=1)
    dilation = cv2.dilate(mask, kernel, iterations=1)
    opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
    
    # difference between input image and opening of the image is tophat
    # difference between input image and closing of the image is tophat
    show("res",res)
    show("mask", mask)
    show("eroded", erosion)
    show('dilate', dilation)
    show("opening", opening)
    show("closing", closing)
    k = cv2.waitKey(5) & 0xFF
    if k==ord("q"):
        break

cv2.destroyAllWindows()
cap.release()
```

# Template matching

```python
img_bgr = cv2.imread("template.jpg")
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

template =  cv2.imread('match.jpg',0)
w,h = template.shape[::-1]

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
threshold = .90
loc = np.where(res>=threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img_bgr, pt, (pt[0]+w, pt[1]+h), (255,0,0), 2)

show("detected", img_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Foreground extraction

```python
img = cv2.imread('test(1).JPG')
mask = np.zeros(img.shape[:2], np.uint8)
bgModel = np.zeros((1,65), np.float64)
fgModel = np.zeros((1,65), np.float64)
print(img.shape)
rect = (0, 0, 500,332)

cv2.grabCut(img, mask, rect, bgModel, fgModel, 5, cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()
```

# Corner Detection

```python
img = cv2.imread('cornerd.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

corners = cv2.goodFeaturesToTrack(gray, 200, .01, 10)
corners = np.int0(corners)

for corner in corners:
    x,y = corner.ravel()
    cv2.circle(img, (x,y), 3, 255, -1)

cv2.imshow('corner', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Feature matching (Homography)

```python
template = cv2.imread('templateH.jpg',0)
image = cv2.imread('imageH.jpg',0)

orb = cv2.ORB_create()

kp1, des1 = orb.detectAndCompute(template, None)
kp2, des2 = orb.detectAndCompute(image, None)

bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck = True)

matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv2.drawMatches(template, kp1, image, kp2, matches[:10], None, flags=2)
plt.imshow(img3)
plt.show()
```

# Harr Cascade Object Detection

```python
face_cascade = cv2.CascadeClassifier("haar-cascade-files/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("haar-cascade-files/haarcascade_eye.xml")
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret,img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (240,0,0), 2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,250,0), 2)
            
    show("img",img)
    k = cv2.waitKey(30) & 0xFF
    if k==ord("q"):
        break
        
cv2.destroyAllWindows()
cap.release()
```
